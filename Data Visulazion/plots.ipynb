```python
import pandas as pd
data = pd.read_csv('/home/mohamed/data-mining-uni-project/Data Preprocessing/WorkoutStats.csv')
print(data)
```

```python
%matplotlib inline

data['Maxpulse'].hist(bins=8)
```

## insights
The histogram shows a peak in 'Maxpulse' values between 120-140 bpm, with the highest frequency around 130-140 bpm (close to 50 occurrences). This suggests that most workouts reach a moderate to high intensity. The distribution tapers off above 160 bpm, indicating fewer extreme efforts.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 2, figsize=(12,12))
index = 0
for i in range(3):
    for j in range(i+1,4):
        ax1 = int(index/2)
        ax2 = index % 2
        axes[ax1][ax2].scatter(data[data.columns[i]], data[data.columns[j]], color='red')
        axes[ax1][ax2].set_xlabel(data.columns[i])
        axes[ax1][ax2].set_ylabel(data.columns[j])
        index = index + 1
```

## insights
For each pair of attributes, we can use a scatter plot to visualize their joint distribution.
